```
path = Path("/content/ecommerce-product-images-multilabel-data")
df = pd.read_csv(path/'train_labels.csv')
df.head()
```

```
dataTypeSeries = df.dtypes
 
print('Data type of each column of Dataframe :')
print(dataTypeSeries)
```

```
df.labelId = df.labelId.astype('str')
```

```
df.labelId = df.labelId.str.replace(',', '')
df.labelId = df.labelId.str.replace('u', '')
df.labelId = df.labelId.str.replace(' ', ';')
df.labelId = df.labelId.str.replace('[', '')
df.labelId = df.labelId.str.replace(']', '')
df.labelId = df.labelId.str.replace('\'', '')
```

```
#df.labelId = df.labelId.str.split(';')
```

```
data = (ImageList.from_df(df, path/"train/train", cols=1, suffix=".jpg")
      .split_by_rand_pct(0.2)
      .label_from_df(label_delim=";")
      .transform(get_transforms(), size=128)
      .databunch(no_check=True)
      .normalize(imagenet_stats))
```

```
data.show_batch()
```

```
data.classes
```

```
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```
arch = models.resnet50
learn = cnn_learner(data, arch, metrics=[accuracy, error_rate])
```

```
learn.lr_find()
learn.recorder.plot()
```
